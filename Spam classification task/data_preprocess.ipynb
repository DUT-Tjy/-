```python
import pandas as pd
from bs4 import BeautifulSoup
import os
```

```python
label_file = 'index.txt'
```

```python
label = pd.read_csv(label_file, sep=' ')
```

```python
label['TYPE'].value_counts()
```

```python
def convert_text_id(label):
    for num in range(len(label)):
        if label['TYPE'][num] == 'ham':
            label['TYPE'][num] = 1
        else:
            label['TYPE'][num] = 0
    return label
```

```python
def convert_id_text(label):
    for num in range(len(label)):
        if label['TYPE'][num] == 1:
            label['TYPE'][num] = 'ham'
        else:
            label['TYPE'][num] = 'spam'
    return label
```

```python
train_text_file = 'train/Data/001/001'
```

```python
count = 0
train_data = pd.DataFrame(columns=('docno','from_data','to','subject', 'text'))
for root, dirs, files in os.walk('train'):
    for file in files:
        with open(os.path.join(root,file), 'r', encoding='gbk', errors='ignore') as f:
            content = f.read()
            soup = BeautifulSoup(content, 'lxml')
            docno = soup.find('docno').string
            from_data = soup.find('from').string
            to = soup.find('to').string
            subject = soup.find('subject').string
            text = soup.find('text').string
            train_data = train_data.append(pd.DataFrame({'docno':str(docno), 'from_data':str(from_data), 'to':str(to), 'subject':str(subject), 'text':str(text)}, index=[count]))
            count = count + 1
            if count % 10000 == 0:
                print(count)
print(count)
```
